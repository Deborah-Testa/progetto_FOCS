```python
datasets = [times_max_year[['university_name','world_rank_times']], shanghai_max_year, cwur_max_year]
```

We define and apply a function that for each value of the world rankings: 
1. if the value has a numeric type or a missing value it returns the value itself
2. if the value is in the range format it returns the mean between the maximum and the minimum value of the interval
3. if the value is in the format of digits preceded by a symbol it returns only the digits

```python
def newrank(el):
    if type(el) != str:
        return el
    else:
        occ1 = re.search('(?P<min>\d+)-(?P<max>\d+)', el)
        occ2 = re.search('^\D?(?P<num>\d+)$', el)
        if occ1:
            return round((float(occ1.group('min')) + float(occ1.group('max')))/2)
        elif occ2:
            return int(occ2.group('num')) 
```

```python
rankings['world_rank_shanghai'] = rankings.pop('world_rank_shanghai').apply(newrank)
rankings['world_rank_times'] = rankings.pop('world_rank_times').apply(newrank)
rankings 
```

We define and apply a function that computes the maximum absolute difference between the three rankings for each university

```python
def absolute_difference(row):
    item1 = row['world_rank_shanghai']
    item2 = row['world_rank_cwur']
    item3 = row['world_rank_times']
    return max(abs(item1 - item2), abs(item1 - item3), abs(item3 - item2))
```

```python
rankings['absolute_difference'] = rankings.apply(absolute_difference, axis = 1)
rankings
```

## 5. Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country.

The rows with missing *'female_male_ratio'* values will be removed, this decision is taken because these rows won't contribute to the final result.  

```python
times_max_year_clean = times_max_year[(times_max_year['female_male_ratio'] != '-') & (times_max_year['female_male_ratio'].notnull())].copy()
times_max_year_clean.head()
```

We define and apply a function that for each row computes the total number of male students and female students

```python
def female_male(row):
    occurrence = re.search('(?P<female>\d+) : (?P<male>\d+)', row['female_male_ratio'])
    if occurrence:
        female = float(occurrence.group('female'))*0.01*row['num_students']
        male = float(occurrence.group('male'))*0.01*row['num_students']
        return [int(round(female)), int(round(male))]
    else:
        return 'Error' 
        #in this way we can control if there are different formats of 'female_male_ratio'
        #that need to be handled 
```

```python
times_max_year_clean['female_male_student'] = times_max_year_clean.apply(female_male, axis = 1)
times_max_year_clean
```

```python
print(f"the number of rows with 'female_male_student' == 'Error' is {len(times_max_year_clean[times_max_year_clean['female_male_student'] == 'Error'])}")
```

It has been verified that the only format of _female_male_ratio_ is 
<code>\d+ : \d+</code>

```python
times_max_year_clean['female'], times_max_year_clean['male'] =  zip(*times_max_year_clean['female_male_student'])
```


## 6. Find the universities where the ratio between female and male is below the average ratio (computed over all universities)

We define and apply a function that selects the first and second term from *female_male_ratio* and computes the ratio

```python
def ratio(row):
    occurrence = re.search('(?P<female>\d+) : (?P<male>\d+)', row['female_male_ratio'])
    return np.float64(occurrence.group('female')) / np.float64(occurrence.group('male'))
```

```python
times_max_year_clean['fm_ratio'] = times_max_year_clean.apply(ratio, axis = 1)
times_max_year_clean
```

We define the *clean_rank* function that checks if the value is in the format of digits preceded by a symbol and it returns only the digits. Then we apply it to the Shanghai and Times datasets.


```python
def clean_rank(el):
    occ = re.search('^\D?(?P<num>\d+)$', el)
    if occ:
        return int(occ.group('num')) 
    else:
        return el
```

We define the *check* function that compares a numeric value with another element:
   1. If the element is a numeric value too, then it checks if they are equal, returning a boolean value
   2. If the element is in a range format, it check if the value is included in the range, returning a boolean value


```python
def check(row):
    if type(row[1])!= str:
        return row[0] == row[1] 
    else:
        occ = re.search('(?P<min>\d+)-(?P<max>\d+)', row[1])
        return (row[0] <= int(occ.group('max'))) & (row[0] >= int(occ.group('min')))
        
```

At first we apply the *check* function to the *world_rank_cwur* and *world_rank_times_clean* columns of the rank_merged_2 dataset

```python
rank_merged_2['check_1']= rank_merged_2[['world_rank_cwur', 'world_rank_times_clean']].apply(check, axis=1)
rank_merged_2.tail()
```

Then we apply the *check* function to the *world_rank_cwur* and *world_rank_shanghai_clean* columns of the rank_merged_2 dataset

```python
rank_merged_2['check_2']= rank_merged_2[['world_rank_cwur', 'world_rank_shanghai_clean']].apply(check, axis=1)
rank_merged_2.tail()
```

Finally we select the rows that satisfy both conditions. As we can see this second solution allows to obtain more results.
